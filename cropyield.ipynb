```python
import json
import matplotlib.pyplot as plt

# Load JSON data from the file
with open("/content/wheatdata.json") as f:
    data = json.load(f)

# Function to extract data and plot relationships
def plot_relationship(data):
    for plot in data:
        plot_name = plot["plot_name"]
        wheat_age_days = [entry["wheat_age_days"] for entry in plot["data"]]
        ndvi = [entry["NDVI"] for entry in plot["data"]]
        temperature = [entry["temperature"] for entry in plot["data"]]
        precipitation = [entry["precipitation"] for entry in plot["data"]]

        # Plot NDVI vs Wheat Age Days
        plt.figure(figsize=(12, 6))
        plt.subplot(1, 3, 1)
        plt.plot(wheat_age_days, ndvi, 'bo-', label="NDVI")
        plt.xlabel("Wheat Age (days)")
        plt.ylabel("NDVI")
        plt.title(f"NDVI vs Wheat Age for {plot_name}")
        plt.legend()
        plt.grid()

        # Plot Temperature vs Wheat Age Days
        plt.subplot(1, 3, 2)
        plt.plot(wheat_age_days, temperature, 'ro-', label="Temperature")
        plt.xlabel("Wheat Age (days)")
        plt.ylabel("Temperature (°C)")
        plt.title(f"Temperature vs Wheat Age for {plot_name}")
        plt.legend()
        plt.grid()

        # Plot Precipitation vs Wheat Age Days
        plt.subplot(1, 3, 3)
        plt.plot(wheat_age_days, precipitation, 'go-', label="Precipitation")
        plt.xlabel("Wheat Age (days)")
        plt.ylabel("Precipitation (mm)")
        plt.title(f"Precipitation vs Wheat Age for {plot_name}")
        plt.legend()
        plt.grid()

        plt.suptitle(f"Relationships for {plot_name}")
        plt.tight_layout()
        plt.show()

plot_relationship(data)
```
