```python
# Install necessary packages
!pip install pandas matplotlib seaborn

# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file from your Google Drive
from google.colab import drive

file_path = '/content/RY_weather_dataset.csv'  # Change this path as necessary

# Load the data
data = pd.read_csv(file_path)

# Filter the data for the district "Durg"
durg_data = data[data['District Name'] == 'Durg']

# Perform correlation analysis
correlation_matrix = durg_data[['Temperature (°C)', 'Precipitation (mm)', 'Rice Yield (Kg per ha)']].corr()
print("Correlation Matrix:")
print(correlation_matrix)

# Plotting the relationships over the years
plt.figure(figsize=(14, 7))

# Line plot for Rice Yield vs Temperature
sns.lineplot(x='Year', y='Temperature (°C)', data=durg_data, marker='o', label='Temperature (°C)')
sns.lineplot(x='Year', y='Rice Yield (Kg per ha)', data=durg_data, marker='o', label='Rice Yield (Kg per ha)')
plt.title('Rice Yield and Temperature Over Years')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()

# Line plot for Rice Yield vs Precipitation
plt.figure(figsize=(14, 7))
sns.lineplot(x='Year', y='Precipitation (mm)', data=durg_data, marker='o', label='Precipitation (mm)')
sns.lineplot(x='Year', y='Rice Yield (Kg per ha)', data=durg_data, marker='o', label='Rice Yield (Kg per ha)')
plt.title('Rice Yield and Precipitation Over Years')
plt.xlabel('Year')
plt.ylabel('Value')
plt.legend()
plt.show()
```
